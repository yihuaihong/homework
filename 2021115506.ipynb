```python
#2
class Solution(object):
    def findItinerary(self, tickets):
        tickets.sort(key=lambda x: (x[0] == "JKF", x[1]) or (x[0], x[1]))

        use = [0] * len(tickets)
        res = ["JFK"]
        final = []

    def findNext(s, tickests, temp, use):
        for i in range(len(tickets)):
            if use[i] != 1 and tickets[i][0] == s:
                temp.append(i)
        return temp

    def backtracking(tickets, s, temp, use):
        if len(res) == len(tickets) + 1:
            final.append(res + [])
            return
        for i in findNext(s, tickets, temp, use):
            if len(final) == 1:
                break
            res.append(tickets[i][1])
            use[i] = 1
            backtracking(tickets, tickets[i][1], [], use)
            res.pop()
            use[i] = 0

    backtracking(tickets, "JFK", [], use)
    return final[0]
```

```python
def anagramSolution1(s1,s2):
    alist = list(s2)  # 复制s2
    pos1 = 0
    stillok = True
    while pos1 < len(s1) and stillok:  # 循环s1的所有字符
        pos2 = 0
        found = False  # 初始化标识符
        while pos2 < len(alist) and not found:  # 与s2的字符逐个对比
            if s1[pos1] == alist[pos2]:
                found = True
            else:
                pos2 = pos2 + 1
        if found:
            alist[pos2] = None  # 找到对应，标记
        else:
            stillok = False  # 没有找到，失败
        pos1 = pos1 + 1
    return stillok

print(anagramSolution1('evil','live'))
```
